```python
from sklearn.preprocessing import MinMaxScaler
from bs4 import BeautifulSoup as bs
from calendar import monthrange
import numpy as np
import ast
import requests
import urllib2
import datetime
import time
import pdb
import re, random
import socket
import os
import json
```

```python
def get_shenzhen_hist(month, year, country="china", city="shenzhen"):
    """
        scraping simple date and temperature data from timeanddate\
        date key of the returned data might have slight time offset with the tick time
    """
    url = "https://www.timeanddate.com/weather/"+country+"/"+city+"/"+"historic?month={0}&year={1}".format(month, year)
    req = urllib2.Request(url)
    resp = urllib2.urlopen(req)
    soup = bs(resp, 'lxml')
    data = soup.findAll("script")[5].text
    data = "["+data[10:-35].encode('ascii', 'ignore')+"]"
    data = json.loads(data)
    return data
```

```python
def get_shenzhen_humidity(data, month, year):
    data = [c['hum'] for c in data[0]['detail']]
    assert len(data)/4 == monthrange(year, month)[1]
    hum_data = list()
    for i in range(len(data)):
        hum_data.extend([data[i]]*6)
    return hum_data
```

```python
def get_shenzhen_wind(data, month, year):
    data = [c['wind'] for c in data[0]['detail']]
    assert len(data)/4 == monthrange(year, month)[1]
    wind_data = list()
    for i in range(len(data)):
        wind_data.extend([data[i]]*6)
    return wind_data
```

```python
def get_shenzhen_temp(data, month, year):
    """
       Fill missing tick point weather data with first value     
    """
    data = [c['temp'] for c in data[0]['temp']]
    days = monthrange(year, month)[1]
    pts_miss = (24 * days)- len(data)
    temp = data[0]
    for _ in range(pts_miss):
        data.insert(0, temp)
    return data
```
