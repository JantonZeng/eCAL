```python
def insert_allStation(connection, station_info): 
    cur = connection.cursor()
    for i in range(len(station_info)): 
        isInterconnection = str(station_info[i]['isInterconnection'])      
        name = str(station_info[i]['name'].encode('utf-8'))
        address = str(station_info[i]['address'].encode('utf-8'))
        isFast = str(station_info[i]['isFast'])      
        stationType = str(station_info[i]['stationType'].encode('utf-8'))   
        stationId = str(station_info[i]['id'])  
        
        longtitude = str(station_info[i]['lng'])
        latitude = str(station_info[i]['lat'])
        try:
            command = (
                """
                INSERT INTO tbl_APP_allStations(
                    isInterconnection, name, address, isFast, stationType, stationId, location
                    ) 
                SELECT %s,%s,%s,%s,%s,%s, ST_SetSRID(ST_MakePoint(%s, %s),4326) 
                    WHERE NOT EXISTS (SELECT 1 FROM tbl_APP_allStations where stationId=%s)
                """)
            cur.execute(command, (isInterconnection, name, address, isFast, 
                                  stationType, stationId, longtitude, latitude, stationId))
        except (Exception, psycopg2.DatabaseError) as error:
            print(error)
    connection.commit()
    return connection
```

```python
def insert_singPile_none(connection, pile_detailedInfo, pillcode):
    """
    This function deals with None information pile.
    pile_detailedInfo = None
    pillcode = u\'unicode'
    """
    cur = connection.cursor()
    pillCode = str(pillcode)
    checkExist = 0
    try:
        command = (
            """
            INSERT INTO tbl_APP_piles(
            pillCode, checkExist
            )
            SELECT %s, %s
            WHERE NOT EXISTS (SELECT 1 FROM tbl_APP_piles where pillCode=%s)
            """
        )
        cur.execute(command, (pillCode, checkExist, pillCode))
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
    connection.commit()
    return connection
```

```python
def insert_pilecode(connection, staID, piles_info):
    """
    Only in this table, it is called pilecode.
    """
    cur = connection.cursor()
    for pile in piles_info:
        pilecode = str(pile['code'])
        try:
            command = (
                """
                INSERT INTO tbl_APP_pileCode(staID, pilecode)
                SELECT %s,%s
                """
                )
            cur.execute(command, (staID, pilecode))
        except (Exception, psycopg2.DatabaseError) as error:
            print(error)
    connection.commit()
    return connection
```

```python
def insert_state_info(connection, pillCode, stateCode, stateName, stateCodeV3, stateNameV3):
    cur = connection.cursor()
    try:
        command = (
            """
            INSERT INTO tbl_app_state(pillCode, stateCode, stateName, stateCodeV3, stateNameV3, time) 
            SELECT %s, %s, %s, %s, %s, Now() 
            WHERE NOT EXISTS (SELECT 1 FROM tbl_app_state where time=Now())
            """)
        cur.execute(command, (pillCode, stateCode, stateName, stateCodeV3, stateNameV3,))
    except (Exception, psycopg2.DatabaseError) as error:
            print(error)
    connection.commit()
    return connection
```
